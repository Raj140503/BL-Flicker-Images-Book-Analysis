```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/BahramJannesar/ChocolateReveiwDataAnalysis/master/datasets/chocolate.csv')
df
```

```python
new_names = {'Unnamed: 0': 'Index',
             'ref': 'ID',
             'company_location': 'Location',
             'review_date': 'Date',
             'country_of_bean_origin': 'Country',
             'specific_bean_origin_or_bar_name': 'Origin',
             'cocoa_percent': 'Percentage',
             'counts_of_ingredients': 'Ingredient Count',
             'cocoa_butter':'Butter'}
```

```python
df.rename(columns=new_names, inplace=True)
df
```

```python
var = df[['first_taste', 'second_taste', 'third_taste', 'fourth_taste']]
var
```

```python
var.groupby(['first_taste', 'second_taste', 'third_taste', 'fourth_taste']).count()
```

```python
a = np.concatenate(('first_taste', 'second_taste', 'third_taste','fourth_taste'), axis=None)
a
```

```python
df_a = pd.DataFrame(a)
df_a.groupby(a).count()
```
