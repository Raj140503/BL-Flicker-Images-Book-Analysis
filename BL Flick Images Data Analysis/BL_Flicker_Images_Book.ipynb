## Data Cleaning

**regex -** Remove Extra Square Brackets

```python
regex = r'^(\d{4})'
```

**.str.extract() -** It is a method to extract substrings from strings in the DataFrame.

```python
extr = df['Date of Publication'].str.extract(regex, expand=False)
extr
```

**pd.to_numeric -** It gives the numerical version of object data type.

```python
df['Date of Publication'] = pd.to_numeric(extr)
df['Date of Publication'].dtype
```

```python
df['Date of Publication'].isnull().sum()/len(df)
```

## Text Formatting

**str.contains() -** To clean the column in one sweep.

```python
pub = df['Place of Publication']
london = pub.str.contains('London')
london[:5]
```

```python
exf = df['Place of Publication']
oxford = exf.str.contains('Oxford')
oxford[:5]
```

**np.where -** It allows to perform element-wise conditional operation on array.

**str.replace -** It replace string within provided text that match with a replacement

```python
df['Place of Publication'] = np.where(london, 'London', np.where(oxford, 'Oxford', pub.str.replace('-', ' ')))
df['Place of Publication'].head()
```
