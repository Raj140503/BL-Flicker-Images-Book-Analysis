```python
import pandas as pd
import numpy as np
```

```python
university_towns = []
with open('/content/university_towns.txt') as file:
    for line in file:
      if '[edit]' in line:    # Remember this line until the next is found
        state = line
      else:
        university_towns.append((state, line))

university_towns[:5]
```

## Convert String DataFrame into pandas

```python
towns_df = pd.DataFrame(university_towns, columns=['State', 'RegionName'])
towns_df.head()
```

```python
def get_city_state(item):
  if '(' in item:
    return item[:item.find('(')]
  elif '[' in item:
    return item[:item.find('[')]
  else:
    return item
```

**applymap() -** function is applied to every element in the dataframe and return new pandas dataframe

```python
towns_df = towns_df.applymap(get_city_state)
towns_df.head()
```

```python
university_count = towns_df.groupby('State').value_counts()
print(university_count)
```
